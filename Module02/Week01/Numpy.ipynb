```python
def maxx(x, y):
    if x >= y:
        return x
    else:
        return y


a = np.array([5, 7, 9, 8, 6, 4, 5])
b = np.array([6, 3, 4, 8, 9, 7, 1])
pair_max = np.vectorize(maxx, otypes=[float])
print(pair_max(a, b))
```

```python
a = np.array([5, 7, 9, 8, 6, 4, 5])
b = np.array([6, 3, 4, 8, 9, 7, 1])
print("Result", np.where(a < b, b, a))
```

```python
import pandas as pd
df = pd.read_csv('./advertising.csv')

data = df.to_numpy()
```

```python
sales_col = data[:, -1]
max_sale = np.max(sales_col)
max_index = np.argmax(sales_col)
print(f"Max value = {max_sale}, index = {max_index}")
```

```python
tv_col = data[:, 0]
mean_tv = tv_col.mean()
mean_tv
```

```python
sales_col = data[:, -1]
sales_counter = np.sum(sales_col >= 20)
sales_counter
```

```python
data[data[:, -1] >= 15][:, 1].mean()
```

```python
mean_newspaper = data[:,2].mean()
data[data[:, 2] > mean_newspaper][:,-1].sum()
```

```python
sales_col = data[:, -1]
A = np.mean(sales_col)
scores = np.where(sales_col > A, "Good", np.where(sales_col < A, "Bad", "Average"))
print(scores[7:10])
```

```python
sales_col = data[:, -1]
mean_val = np.mean(sales_col)
A = sales_col[np.abs(sales_col - mean_val).argmin()]
scores = np.where(sales_col > A, "Good", np.where(sales_col < A, "Bad", "Average"))
print(scores[7:10])
```
