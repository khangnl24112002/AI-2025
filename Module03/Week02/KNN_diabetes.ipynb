```python
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import r2_score
import matplotlib.pyplot as plt
```

```python

diabetes_X, diabetes_y = datasets.load_diabetes(return_X_y=True)
X_train, X_test, y_train, y_test = train_test_split(
    diabetes_X,
    diabetes_y,
    test_size=0.2,
    random_state=42,
)
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# 2. Thử nghiệm các giá trị K khác nhau
neighbors = np.arange(1, 41) # Thử K từ 1 đến 40
r2_scores = []

for k in neighbors:
    # Khởi tạo và huấn luyện mô hình KNN Regressor với giá trị k hiện tại
    knn_regressor = KNeighborsRegressor(n_neighbors=k)
    knn_regressor.fit(X_train, y_train)
    
    # Dự đoán trên tập test
    y_pred = knn_regressor.predict(X_test)
    
    # Tính R2 score và lưu lại
    score = r2_score(y_test, y_pred)
    r2_scores.append(score)
    print(f"Với K = {k}, R2 Score = {score:.4f}")

# 3. Trực quan hóa kết quả
plt.figure(figsize=(10, 6))
plt.plot(neighbors, r2_scores, marker='o', linestyle='-')
plt.title('Ảnh hưởng của K đến R2 Score')
plt.xlabel('Số láng giềng (K)')
plt.ylabel('R2 Score')
plt.grid(True)
# Tìm và đánh dấu giá trị K tốt nhất
best_k = neighbors[np.argmax(r2_scores)]
best_score = max(r2_scores)
plt.axvline(x=best_k, color='r', linestyle='--', label=f'K tốt nhất = {best_k} (R2 = {best_score:.4f})')
plt.legend()
plt.show()
```
